```python
import random
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
from nltk.corpus import stopwords
import spacy

nlp = spacy.load('en_core_web_sm')
```

```python
def tokenize_and_lemmatize(text):
    tokens = nltk.word_tokenize(text.lower())
    tokens = [token for token in tokens if token not in string.punctuation]
    tokens = [token for token in tokens if token not in stopwords.words('english')]
    lemmatizer = nltk.stem.WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return tokens
```

```python
def generate_response(user_input):
    bot_response = ''
    sent_tokens.append(user_input)

    TfidfVec = TfidfVectorizer(tokenizer=tokenize_and_lemmatize, stop_words='english')
    tfidf = TfidfVec.fit_transform(sent_tokens)

    vals = cosine_similarity(tfidf[-1], tfidf)
    idx = vals.argsort()[0][-2]
    flat = vals.flatten()
    flat.sort()

    req_tfidf = flat[-2]

    if req_tfidf == 0:
        bot_response = bot_response + "I am sorry! I don't understand you"
        return bot_response
    else:
        bot_response = bot_response + sent_tokens[idx]
        return bot_response
```

```python
flag = True
print("Bot: Hello! I am a chatbot. I will try to respond to you. If you want to exit, type Bye!")

while flag:
    user_input = input('You: ')
    user_input = user_input.lower()

    if user_input != 'bye':
        if user_input == 'thanks' or user_input == 'thank you':
            flag = False
            print("Bot: You are welcome!")
        else:
            if generate_response(user_input) == "I am sorry! I don't understand you":
                print("Bot: ", generate_response(user_input))
                sent_tokens.remove(user_input)
            else:
                print("Bot: ", generate_response(user_input))
                sent_tokens.remove(user_input)
    else:
        flag = False
        print("Bot: Goodbye! Take care.")
```
